```python
# type: ignore
import numpy as np
import colour
import plotly.graph_objs as go

shape = colour.SpectralShape(
    colour.SPECTRAL_SHAPE_DEFAULT.start,
    colour.SPECTRAL_SHAPE_DEFAULT.end,
    10
)

cmfs = colour.MSDS_CMFS["CIE 1931 2 Degree Standard Observer"]

XYZ = colour.volume.XYZ_outer_surface(cmfs.copy().align(shape), colour.SDS_ILLUMINANTS["D65"])
oklab = colour.XYZ_to_Oklab(XYZ)
X, Y, Z = colour.utilities.tsplit(oklab)

with open("oklab-gamut-d65.txt", mode="w", encoding="utf8") as file:
    for L, a, b in oklab:
        file.write(f"{L}, {a}, {b}\n")

XYZtoo = colour.volume.XYZ_outer_surface(cmfs.copy().align(shape))
oklabtoo = colour.XYZ_to_Oklab(XYZtoo)
X2, Y2, Z2 = colour.utilities.tsplit(oklabtoo)

with open("oklab-gamut-e.txt", mode="w", encoding="utf8") as file:
    for L, a, b in oklabtoo:
        file.write(f"{L}, {a}, {b}\n")

trace = go.Scatter(x=Y, y=Z, marker=dict(color="#0040ff"))
trace2 = go.Scatter(x=Y2, y=Z2, marker=dict(color="#ff4020"))
layout = go.Layout(width=1000, height=1000, margin=dict(l=0, r=0, b=0, t=0))
figure = go.Figure(data=[trace, trace2], layout=layout)
figure.show()
```
